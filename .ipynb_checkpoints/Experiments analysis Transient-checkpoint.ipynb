```python
import glob
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing import event_accumulator
```

```python
def plot_event(event, lr, latent_dim):
    ea = event_accumulator.EventAccumulator(event)
    ea.Reload() 
    ea.Tags()
    metrics = ['Reconstruction_Loss']
    fig, axs = plt.subplots(nrows=1, ncols=len(metrics), figsize=(20,3))
    fig.suptitle('Learning rate: {}, Latent Dimension:Â {} \n'.format(lr, latent_dim), fontsize=20)
    for i, metric in enumerate(metrics):
        axs.set_title(metric)
        train = pd.DataFrame(ea.Scalars('{}_train'.format(metric))).value.values
        val = pd.DataFrame(ea.Scalars('{}_val'.format(metric))).value.values
        axs.plot(np.arange(len(train)), train, c='k', label='train')
        axs.plot(np.arange(len(val)), val, c='b', label='val')
        axs.set_xlabel('Epoch')
        axs.set_ylabel(metric)
        axs.grid(True)
    fig.tight_layout()
    fig.subplots_adjust(top=0.8)
    plt.show()
```

```python
lrs = [5e-3, 1e-3, 5e-4, 1e-4, 5e-5, 1e-5]
latent_dims = [8, 16, 32, 64, 128]
hierClass = 'Transient'
outlier = 'Bogus'
folds = 5

experiments = {}
for latent_dim in latent_dims:
    for lr in lrs:
        path = 'results/latentdim{}_lr{}_{}_{}'.format(latent_dim, str(lr),
                hierClass, outlier)
        losses_val = []
        losses_test = []
        for fold in range(folds):
            #event_path = glob.glob('{}_r{}/event*'.format(path, repetition))[0]
            #plot_event(event_path, temperature, alpha)
            try:
                metrics_val_path = '{}_fold{}/metrics_val.json'.format(path, fold)
                loss_val = json.load(open(metrics_val_path))
                losses_val.append(loss_val['Reconstruction Loss'])

                metrics_test_path = '{}_fold{}/metrics_test.json'.format(path, fold)
                loss_test = json.load(open(metrics_test_path))
                losses_test.append(loss_test['Reconstruction Loss'])
            except:
                pass
        
        mean_loss_val = np.mean(losses_val)
        std_loss_val = np.std(losses_val)
        experiments['latentdim{}_lr{}_val'.format(latent_dim, str(lr))] = (mean_loss_val, std_loss_val)
        mean_loss_test = np.mean(losses_test)
        std_loss_test = np.std(losses_test)
        experiments['latentdim{}_lr{}_test'.format(latent_dim, str(lr))] = (mean_loss_test, std_loss_test)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=6, sharey=True, sharex=True, figsize=(18,4))
i = 0
for lr in lrs:
    plt.figure()
    for latent_dim in latent_dims:
        exp_name = 'latentdim{}_lr{}_val'.format(latent_dim, str(lr))
        try:
            mean_f1 = experiments[exp_name][0]
            std_f1 = experiments[exp_name][1]
            axs[i].set_title('Learning Rate: {}'.format(str(lr)))
            axs[i].errorbar(latent_dim, mean_f1, yerr=std_f1, fmt='o')
            axs[i].grid(True)
        except:
            pass
    i+=1
plt.setp(axs[:], xlabel='latent_dim')
plt.setp(axs[:], ylabel='Reconstruction')
fig.savefig('results/rec_vs_lrs_and_latentdim_val_{}.png'.format(hierClass))
fig.show()
```

```python
lr = 5e-4
latent_dim = 64
fold = 0
path = 'results/latentdim{}_lr{}_{}_{}'.format(latent_dim, str(lr), hierClass, outlier)
event_path = glob.glob('{}_fold{}/event*'.format(path, fold))[0]
plot_event(event_path, lr, latent_dim)
```

```python
from model import Autoencoder
from preprocess import get_dataloader
import torch, os

#directory of the best model chosing
directory = '{}_fold{}'.format(path, fold)

#Restoring the best parameters found.
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = Autoencoder(latent_dim).to(device)
weights_dir = os.path.join(directory, 'weights')
state_dict = torch.load('{}/trained_parameters.pth'.format(weights_dir))
model.load_state_dict(state_dict)
```

```python
from joblib import load


scaler_filename = '{}/scaler.save'.format(directory)
scaler = load(scaler_filename)  
dataloader_test, _, class_dict = get_dataloader(128, 'train', hierClass, 
                                 outlier, upsample=False, scaler=scaler, fold=fold)
```

```python
from sklearn.manifold import TSNE

latents = []
labels= []
model.eval()
with torch.no_grad():
    for _, x, y, _, _ in dataloader_test:
        x = x.float().to(device)
        x_hat, z = model(x)
        latents.append(z.detach().cpu())
        labels.append(y.cpu())
labels = torch.cat(labels).numpy().reshape(-1,)
latents = torch.cat(latents).numpy()
x_embedded = TSNE(n_components=2).fit_transform(latents)
```

```python
plt.figure(figsize=(8,8))
plt.title('TSNE for train points (AE)')
cmap = plt.get_cmap('jet', 4)
for key, value in class_dict.items():
    plt.scatter(x_embedded[:, 0][labels==value], x_embedded[:, 1][labels==value], 
                s=20, alpha=1, marker='.', label=key)
plt.legend()
plt.grid(True)
```
